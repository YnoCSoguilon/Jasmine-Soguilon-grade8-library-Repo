```python
import firebase_admin
from firebase_admin import credentials, db

# Use the uploaded key
cred = credentials.Certificate("firebase_key.json")

# Replace with your Realtime Database URL from Firebase console
# Example: "https://your-project-id-default-rtdb.asia-southeast1.firebasedatabase.app"
firebase_admin.initialize_app(cred, {
    "databaseURL": "https://jasmine-soguilon-grade8-lib-db-default-rtdb.asia-southeast1.firebasedatabase.app/"
})
```

```python
# Assuming books_data is already loaded from GitHub
ref = db.reference("/")   # root of the database
ref.set(books_data)       # this writes {"books": [...]} to the DB

print("Data uploaded.")
```

```python
# Read everything from the root of the database
ref = db.reference("/")        # root path
data_from_db = ref.get()       # this returns a Python dict

print(data_from_db)
```

```python
books = data_from_db["books"]  # list of book dictionaries

for book in books:
    print(book["title"])
```

```python
for book in books:
    title = book["title"]
    available = book["details"]["available"]
    print(f"Title: {title}, Available: {available}")
```

```python
total_copies = 0

for book in books:
    total_copies += book["details"]["copies"]

print("Total number of books in the library:", total_copies)
```
